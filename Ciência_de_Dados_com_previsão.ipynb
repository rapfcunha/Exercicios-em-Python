```python
#Passo a passo:
#Passo 1: Entender o desafio proposto
#Passo 2: Entendimento da Ã¡rea/empresa
#Passo 3: ExtraÃ§Ã£o/obtenÃ§Ã£o dos dados
import pandas as pd
tabela = pd.read_csv('/content/barcos_ref.csv')
display(tabela)
```

```python
#Passo 4: Ajuste de dados (tratamento/limpeza)
print(tabela.info())

#Tabela totalmente preenchida
```

```python
#Passo 5: AnÃ¡lise exploratÃ³ria
correlacao = tabela.corr()[['Preco']]
display(correlacao)
import seaborn as sns
import matplotlib.pyplot as plt

#cria o grÃ¡fico
sns.heatmap(correlacao, cmap='Blues', annot=True)
#exibe o grÃ¡fico
plt.show()
```

```python
  #Passo 6: Modelagem + Algoritmos
  #dividir a base em x e y
  y = tabela['Preco']
  x = tabela.drop('Preco', axis=1)

  #train test split
  from sklearn.model_selection import train_test_split
  x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
#Importar a inteligÃªncia artificial
#regressÃ£o linear e Ã¡rvore de decisÃ£o
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
#cria a inteligÃªncia
modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()
#treinar a inteligÃªncia artificial 
modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvoredecisao.fit(x_treino, y_treino)
```

```python
#Passo 7: InterpretaÃ§Ã£o do resultado

#escolher o melhor modelo -> RÂ²
from sklearn.metrics import r2_score
previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
print(r2_score(y_teste, previsao_regressaolinear))
print(r2_score(y_teste, previsao_arvoredecisao))
```

```python
#visualizar as previsÃµes

tabela_auxiliar = pd.DataFrame()
tabela_auxiliar['y_teste'] = y_teste
tabela_auxiliar['ArvoreDecisao'] = previsao_arvoredecisao
tabela_auxiliar['RegressaoLinear'] = previsao_regressaolinear

sns.lineplot(tabela_auxiliar)
plt.show()
```

```python
#fazer novas previsÃµes
tabela_nova = pd.read_csv('/content/novos_barcos.csv')
display(tabela_nova)

previsao = modelo_arvoredecisao.predict(tabela_nova)
print(previsao)
```
